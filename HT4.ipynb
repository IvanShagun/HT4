## Задание 3 
*Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br>
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.*

```python
cor_m = round(df.drop(columns=['id', 'date', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
                     ).corr(method='spearman'), 2)
cor_m[np.abs(cor_m) < 0.3] = 0
```

```python
plt.figure(figsize=(12, 8))
plt.title('Feature correlation matrix', fontsize=20);

sns.heatmap(cor_m, annot=True, linewidths=.5, cmap='coolwarm');
```

##### Вывод:<br>
Мы можем наблюдать заметную зависимость цены дома от конструкции здания, жилой площади, а также от площади неподвального помещения.<br>Площадь неподвального помещения рассматривать не стоит, т.к. она схожа с общей жилой площадью.<br>Умренная зависимость имеется от этажности, количества спален и ванных комнат.

```python
df[df['grade'] >= 11]
```

```python
grd = df.groupby('grade')['price'].mean().reset_index()
grd
```

```python
plt.bar(grd['grade'], grd['price'])

plt.title('Price - building condition & design ratio', fontsize=15)
plt.xlabel('Building condition & design grade')
plt.ylabel('Mean price')
plt.xticks(range(0, 14));
```

Средняя цена домов растет по мере улучшения качества конструкции и дизайна. Цена на дома с показателями 12 и 13 значительно выше по отношению к более низкими показателями. Данный рост цен может быть связан с применением в строительстве более дорогих и качественных стройматериалов и дорогой декорацией.

```python
plt.figure(figsize=(10,6))

sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');

plt.xlabel('Price (mln)', fontsize=13)
plt.ylabel('Living square (ft)', fontsize=13)
plt.grid();
```

На данном графике мы видим, что цена растет с увеличением жилой площади. В низком ценовом диапазоне данная зависимость выражена более ярко. 

```python
btr = df.groupby('bathrooms')['price'].mean().reset_index()
btr
```

```python
df.loc[df['bathrooms'] >= 6]
```

```python
plt.figure(figsize=(10, 6))

plt.plot(btr['bathrooms'], btr['price'])

plt.title('Mean price - bathrooms ratio', fontsize=15)
plt.xlabel('Number of bathrooms')
plt.ylabel('Mean price')
plt.xticks(btr['bathrooms'], rotation=600)
plt.grid()
```

Наболюдается плавное увеличение средней стоимости в зависимости от увеличения количества ванных комнат. Выбросы схожи с домами типа таунхаус.
